## Dropping Unnecessary Cols

```python
df.columns.values
```

## Top 10 Countries w.r.t. total Population

```python
# Filter total population data
total_pop_data = df[df['Series Code'] == 'SP.POP.TOTL']

# Sort by 2022 population
total_pop_sorted = total_pop_data.sort_values(by='2022', ascending=False)

# Top 10 countries
total_top_countries = total_pop_sorted.head(10)

print("Top 10 countries by total population")
print(total_top_countries['Country Name'])
```

# Bar Graph Plotting

## Top Ten Countries with Total Population in 2016 & 2022

```python
# Creating bar graph
# 2016 Yr
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x = '2016', y = 'Country Name', data = total_top_countries, palette = 'dark')
plt.title("Top Ten Countries of Total Population (2016)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

# 2022 Yr
plt.figure(figsize=(15, 6))
plt.subplot(2,2,2)
sns.barplot(x = '2022', y = 'Country Name', data = total_top_countries, palette = 'dark')
plt.title("Top Ten Countries of Total Population (2022)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

## Top 10 Countries with Highest Male Population 

```python
# Filter data
male_pop_data = df[df["Series Code"] == "SP.POP.TOTL.MA.IN"]

# Sort Male Pop. data for 2022
male_pop_sorted = male_pop_data.sort_values(by="2022", ascending=False)

# Top Countries with Highest M. Pop 2022
male_top_ten_countries = male_pop_sorted.head(10)
print("Top ten countries of male population")
print(male_top_ten_countries[['Country Name']] )


# Bar Plotting
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="2022", y="Country Name", data=male_top_ten_countries, palette="rocket")
plt.title("Top Ten Countries with Highest Male Population (2022)",size=10)
plt.xlabel("Male Population",size=10)
plt.ylabel("Country",size=10)
plt.show()
```

## Top 10 Countries with Highest Female Population 

```python
# Filter data
female_pop_data = df[df["Series Code"] == "SP.POP.TOTL.FE.IN"]

# Sort Female Pop. data for 2022
female_pop_sorted = female_pop_data.sort_values(by="2022", ascending=False)

# Top Countries with Highest F. Pop 2022
female_top_ten_countries = female_pop_sorted.head(10)
print("Top ten countries of female population")
print(female_top_ten_countries[['Country Name']] )


# Bar Plotting
plt.figure(figsize=(15, 6))
plt.subplot(2,2,2)
sns.barplot(x="2022", y="Country Name", data=female_top_ten_countries, palette="rocket")
plt.title("Top Ten Countries with Highest Female Population (2022)",size=10)
plt.xlabel("Female Population",size=10)
plt.ylabel("Country",size=10)
plt.show()
```

## Top Countries with Lowest Male Population

```python
male_lowest_countries = male_pop_sorted.tail(10)
print("Top ten countries of lowest male population")
print(male_lowest_countries[['Country Name']] )


# Bar Graph Plot
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="2022", y="Country Name", data=male_lowest_countries, palette="rocket")
plt.title("Top Ten Countries with Lowest Male Population (2022)",size=10)
plt.xlabel("Male Population",size=10)
plt.ylabel("Country",size=10)
plt.show()
```

## Top Countries with Lowest Female Population

```python
female_lowest_countries = female_pop_sorted.tail(10)
print("Top ten countries of lowest male population")
print(female_lowest_countries[['Country Name']] )


# Bar Graph Plot
plt.figure(figsize=(15, 6))
plt.subplot(2,2,2)
sns.barplot(x="2022", y="Country Name", data=female_lowest_countries, palette="rocket")
plt.title("Top Ten Countries with Lowest Female Population (2022)",size=10)
plt.xlabel("Female Population",size=10)
plt.ylabel("Country",size=10)
plt.show()
```

# Stacked Bar Plot

## Top 10 Countries with Male and Female Populations (2022)

```python
# Merge male and female population data on 'Country Name'
merged_data = pd.merge(male_pop_data, female_pop_data, on="Country Name", suffixes=("_male", "_female"))
merged_data
```

```python
# Calculate the total population for each country (male + female)
merged_data["Total Population"] = merged_data["2022_male"] + merged_data["2022_female"]
merged_data.head()
```

```python
# Sort data based on total population in descending order
sorted_data = merged_data.sort_values(by="Total Population", ascending=False)

# Select the top 10 countries with the highest total population
top_10_countries = sorted_data.head(10)
```

```python
# Stacked Bar Graph 
plt.figure(figsize=(12, 6))

sns.barplot(x="Country Name", y="2022_female", data=top_10_countries, color="#ff028d", label="Female Population")
sns.barplot(x="Country Name", y="2022_male", data=top_10_countries, bottom=top_10_countries["2022_female"], color="blue", label="Male Population")
plt.title("Top 10 Countries with Male and Female Populations (2022)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45, ha="right")
plt.show()
```

## Bottom 10 Countries with Male and Female Populations (2022)

```python
# Select the top 10 countries with the highest total population
bottom_10_countries = sorted_data.tail(10)
```

```python
# Create the stacked bar plot
plt.figure(figsize=(12, 6))

sns.barplot(x="Country Name", y="2022_female", data=bottom_10_countries, color="#ff028d", label="Female Population")
sns.barplot(x="Country Name", y="2022_male", data=bottom_10_countries, bottom=bottom_10_countries["2022_female"], color="blue", label="Male Population")
plt.title("Bottom 10 Countries with Male and Female Populations (2022)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45, ha="right")
plt.show()
```
