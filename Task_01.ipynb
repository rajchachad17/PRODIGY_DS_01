## Top 10 Countries w.r.t. total Population

```python
# Filter total population data
total_pop_data = df[df['Series Code'] == 'SP.POP.TOTL']

# Sort by 2022 population
total_pop_sorted = total_pop_data.sort_values(by='2022', ascending=False)

# Top 10 countries
total_top_countries = total_pop_sorted.head(10)

print("Top 10 countries by total population")
print(total_top_countries['Country Name'])
```

## Top 10 Countries with Highest Male & Female Population 

```python
# Filter data
male_pop_data = df[df["Series Code"] == "SP.POP.TOTL.MA.IN"]
female_pop_data = df[df["Series Code"] == "SP.POP.TOTL.FE.IN"]

# Sort data for 2022
male_pop_sorted = male_pop_data.sort_values(by="2022", ascending=False)
female_pop_sorted = female_pop_data.sort_values(by="2022", ascending=False)

# Top 10 Countries
male_top_ten_countries = male_pop_sorted.head(10)
female_top_ten_countries = female_pop_sorted.head(10)

print("Top 10 Countries by Male Population")
print(male_top_ten_countries[['Country Name']])

print("\nTop 10 Countries by Female Population")
print(female_top_ten_countries[['Country Name']])


# Creating Bar Graphs
plt.figure(figsize=(16,5))

# Male Population
plt.subplot(1,2,1)
sns.barplot(
    x='2022',
    y='Country Name',
    data=male_top_ten_countries,
    palette='mako'
)
plt.title("Top 10 Countries by Male Population (2022)", fontsize=14, fontweight='bold')
plt.xlabel("Male Population", fontsize=12)
plt.ylabel("Country", fontsize=12)

# Female Population
plt.subplot(1,2,2)
sns.barplot(
    x='2022',
    y='Country Name',
    data=female_top_ten_countries,
    palette='rocket'
)
plt.title("Top 10 Countries by Female Population (2022)", fontsize=14, fontweight='bold')
plt.xlabel("Female Population", fontsize=12)
plt.ylabel("Country", fontsize=12)

plt.tight_layout()
plt.show()
```

## Top Countries with Lowest Male & Female Population

```python
# Bottom 10 Countries by Male & Female Population (2022)

male_lowest_countries = male_pop_sorted.tail(10)
female_lowest_countries = female_pop_sorted.tail(10)

plt.figure(figsize=(16,5))

# Male Population
plt.subplot(1,2,1)
sns.barplot(
    x='2022',
    y='Country Name',
    data=male_lowest_countries,
    palette='mako'
)
plt.title("Bottom 10 Countries by Male Population (2022)", fontsize=14, fontweight='bold')
plt.xlabel("Male Population", fontsize=12)
plt.ylabel("Country", fontsize=12)

# Female Population
plt.subplot(1,2,2)
sns.barplot(
    x='2022',
    y='Country Name',
    data=female_lowest_countries,
    palette='rocket'
)
plt.title("Bottom 10 Countries by Female Population (2022)", fontsize=14, fontweight='bold')
plt.xlabel("Female Population", fontsize=12)
plt.ylabel("Country", fontsize=12)

plt.tight_layout()
plt.show()
```

# Stacked Bar Plot

## Top 10 Countries with Male and Female Populations (2022)

```python
# Merge male and female population data on 'Country Name'
merged_data = pd.merge(male_pop_data, female_pop_data, on="Country Name", suffixes=("_male", "_female"))
merged_data
```

```python
# Calculate the total population for each country (male + female)
merged_data["Total Population"] = merged_data["2022_male"] + merged_data["2022_female"]
merged_data.head()
```

```python
# Sort data based on total population in descending order
sorted_data = merged_data.sort_values(by="Total Population", ascending=False)

# Select the top 10 countries with the highest total population
top_10_countries = sorted_data.head(10)
```

```python
# Creating Stacked Bar Graph

plt.figure(figsize=(10,6))

plt.bar(
    top_10_countries["Country Name"],
    top_10_countries["2022_female"],
    color="hotpink",
    label="Female Population"
)

plt.bar(
    top_10_countries["Country Name"],
    top_10_countries["2022_male"],
    bottom=top_10_countries["2022_female"],
    color="steelblue",
    label="Male Population"
)

plt.title("Top 10 Countries by Male and Female Population (2022)", fontsize=14, fontweight='bold')
plt.xlabel("Country", fontsize=12)
plt.ylabel("Population", fontsize=12)

plt.xticks(rotation=45, ha='right')
plt.legend()

plt.tight_layout()
plt.show()
```

## Bottom 10 Countries with Male and Female Populations (2022)

```python
# Select the top 10 countries with the highest total population
bottom_10_countries = sorted_data.tail(10)
```

```python
# Creating Grouped Bar Chart

plt.figure(figsize=(10,6))

x = range(len(bottom_10_countries))
width = 0.4

plt.bar(
    [i - width/2 for i in x],
    bottom_10_countries["2022_male"],
    width=width,
    color="steelblue",
    label="Male Population"
)

plt.bar(
    [i + width/2 for i in x],
    bottom_10_countries["2022_female"],
    width=width,
    color="hotpink",
    label="Female Population"
)

plt.xticks(x, bottom_10_countries["Country Name"], rotation=45, ha='right')

plt.title("Bottom 10 Countries by Male and Female Population (2022)", fontsize=14, fontweight='bold')
plt.xlabel("Country", fontsize=12)
plt.ylabel("Population", fontsize=12)

plt.legend()

plt.tight_layout()
plt.show()
```
